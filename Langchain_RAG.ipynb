```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
from langchain_community.vectorstores import FAISS
```

```python
!pip install faiss-cpu
```

```python
db = FAISS.from_documents(documents=data, embedding=embeddings)
```

```python
retriever = db.as_retriever()
```

```python
result = retriever.invoke("Please do provide me the links related to the React.js")
print('Links Bro', result)
```

```python
from langchain.chains import RetrievalQA
```

```python
retrievalQA = RetrievalQA.from_llm(llm=llm, retriever=retriever)
retrievalQA('Please do provide me the links related to the React.js')
```

```python
store = retrievalQA('Please do provide me multiple links related to the React.js')
print('store', store)
```
